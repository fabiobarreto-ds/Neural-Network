```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from tensorflow.keras.preprocessing import image
import warnings
warnings.filterwarnings("ignore")
```

```python
model = Sequential()

#1ª camada
model.add(Conv2D(64, (3,3), input_shape=(64,64,3), activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))

#2ª camada
model.add(Conv2D(64, (3,3), input_shape=(64,64,3), activation="relu"))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(units=128, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(units=128, activation="relu"))
model.add(Dropout(0.2))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
gerador_treinamento = ImageDataGenerator(rescale=1./255, rotation_range=7, horizontal_flip=True, shear_range=0.2, height_shift_range=0.07, zoom_range=0.2)
gerador_teste = ImageDataGenerator(rescale=1./255)
```

```python
base_treinamento = gerador_treinamento.flow_from_directory("C:\\Users\\user\\Data Science\\banco_dados\\datasets\\dataset\\dataset\\training_set", 
                                                           target_size=(64,64),
                                                           batch_size=32,
                                                           class_mode="binary")
```

```python
base_teste = gerador_teste.flow_from_directory("C:\\Users\\user\\Data Science\\banco_dados\\datasets\\dataset\\dataset\\test_set", 
                                                           target_size=(64,64),
                                                           batch_size=32,
                                                           class_mode="binary")
```

```python
model.fit(base_treinamento, 
          steps_per_epoch=len(base_treinamento)//32,
          epochs=1000,
          validation_data = base_teste,
          validation_steps= len(base_teste)//32)
```
