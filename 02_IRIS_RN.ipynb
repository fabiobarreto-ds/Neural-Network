```python
df = pd.read_csv("iris.csv")
X = df.drop(columns=["species"])
y = df["species"]
le = LabelEncoder()
y = le.fit_transform(y)

#====================================================================================================
def criarRede(optimizer='rmsprop', init='glorot_uniform'):
    k.clear_session()
    rn = Sequential([
                #1ª camada oculta
               tf.keras.layers.Dense(units=4, activation = "relu", kernel_initializer = "random_uniform", input_dim=4), # input_dim = nº de colunas
               tf.keras.layers.Dropout(0.2),
               #Camada de saída
               tf.keras.layers.Dense(units= 3, activation = 'softmax')])
    

    rn.compile(optimizer = optimizer, loss = "sparse_categorical_crossentropy", metrics = ['accuracy'])


    return rn
#====================================================================================================
kc = KerasClassifier(model=criarRede)

pipe = Pipeline([('ss', StandardScaler()),
                 ('rn', kc)])

# Opitmização de Hiperparêmetros
params = {"rn__optimizer": ["adam", "sgd"],
          "rn__batch_size": [10, 15],
          "rn__epochs":[1000]} 
#====================================================================================================
grid_search = GridSearchCV(estimator=pipe,
                           param_grid=params,
                           cv=10)

grid_search = grid_search.fit(X, y)
y_proba = grid_search.predict_proba(X)
```

# **Classificação de um novo registro:**

```python
df = pd.read_csv("iris.csv")
X = df.drop(columns=["species"])
y = df["species"]
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
rn = Sequential([
                #1ª camada oculta
               tf.keras.layers.Dense(units=4, activation = "relu", kernel_initializer = "random_uniform", input_dim=4), # input_dim = nº de colunas
               tf.keras.layers.Dropout(0.2),
               #Camada de saída
               tf.keras.layers.Dense(units= 3, activation = 'sigmoid')])
    

rn.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ['accuracy'])
```

```python
rn.fit(X,y, batch_size= 10, epochs=1000)
```

```python
novo_registro = np.array([[4.5, 3.2, 1.3, 0.2]])
```

```python
y_preditivo = rn.predict(novo_registro)
y_preditivo 
```
